# **Aprendizaje de regular expresions**

Regex nos permite encontrar y realiar busquedas en un cadena de texto lo que nos permite encontrar ciertos patrones o limpiar ciertos textos.

```python
import re
import numpy as np
```

```python
text = "today was a good day"

re.search("was", text)
```

```python
text = "ale es bueno, ale es aasa, ale adora a all"
re.split("ale",text)
re.findall("ale", text)
 
```

```python
# El corchete nos permite realizar busqueda entre varias opcionen


grades="ABCBBAABBABBCBCBBAD"

re.findall("[AB]", grades)

# Tambien podemos poner rando por ejemplo [a-z] (todos las letras minusculas)

re.findall("[A][B-D]", grades) # buscar todos los patrones tales que comiencen con una A y luego un B,C o D
```

```python
re.findall("AB|CB", grades)
```

```python
re.findall("[^A]",grades) ## Todas excepto A

#Tener presente que no es lo mismo que escribir:
re.findall("^A",grades) # Todos los que inician con A
```

```python
re.findall("A{2,10}",grades) #Todas las veces que hubo una A seguida de maximo 10 A
```

```python
re.findall("A{1,1}A{1,1}",grades) #Todas las veces que hubo [AA]

re.findall("A{2,2}",grades)
```

```python
re.findall("[a-zA-Z]{1,100000}\[edit\]",wiki)
```

```python
re.findall("[\w]{1,100000}\[edit\]",wiki)

# \w indica todos los caracter alphanumericos 
```

```python
re.findall("[\w]*\[edit\]",wiki)

# El caracter * indica que el patron puede aparcer todas las veces sin importar el numero
```

```python
re.findall("[\w ]*\[edit\]",wiki)
```

```python
def l2_dist(a, b):
    result = ((a - b) * (a - b)).sum()
    result = result ** 0.5
    return result 
l2_dist(np.reshape(a, (20 * 20)), np.reshape(b, (20 * 20, 1)))
```
