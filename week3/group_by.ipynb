# **Groupby()**

La funcion groupby() nos permite iterar sobre cada columna de una manera eficiente

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("census.csv")

df = df[df["SUMLEV"]==50]
df.head()
```

```python
%%timeit -n 3

for state in df["STNAME"].unique():
    avg = np.average(df.where(df["STNAME"]==state).dropna()["CENSUS2010POP"])
    print("State: " + state + "avg POP2010: " + str(avg.round(0)))


```

```python
%%timeit -n 3

for group, frame in df.groupby("STNAME"):
    avg = np.average(frame["CENSUS2010POP"])

    print("Estado: " + group + " | Promedio: " + str(avg))
```

** Cuando interamos con groupby solo tomamos una columna
