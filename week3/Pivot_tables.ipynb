# **Pivot Tables**

Una pivot table, tambien considerada tabla dinamica es un tido de dataframe 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cwurData.csv')
# Deseamos rankear las universidades por calidad de educación por los tanto de 0 a 100 seran top tier, 
# de 201 a 300 seran mid tier y de 300 a 1000 seran low tier

df

df['Rank_Level'] = df['world_rank'].apply(lambda x: 'Top Tier' if x < 100 else ('Mid Tier' if x <= 200 and x >= 100 else 'Low Tier'))
df
```

```python
df.pivot_table(values="score",index="country",columns="Rank_Level",aggfunc=[np.mean])
```
