```python
antioquia = df_final.loc["ANTIOQUIA"]
antioquia_conservador = antioquia[antioquia["codigo_partido"]=="Partido conservador"]
antioquia_conservador
```

```python
lista_municipio = []
lista_votos = []
for group, frame in antioquia_conservador.groupby("municipio"):
    votos = np.sum(frame["votos"])
    lista_municipio.append(group)
    lista_votos.append(votos)

votos_conservador = pd.DataFrame(list(zip(lista_municipio, lista_votos)),
               columns =['Municipio', 'Votos_conservador'])
votos_conservador.sort_values(by=["Votos_conservador"],ascending=False)

```

```python
lista_municipio = []
lista_votos = []
for group, frame in antioquia.groupby("municipio"):
    votos = np.sum(frame["votos"])
    lista_municipio.append(group)
    lista_votos.append(votos)

votos_antioquia= pd.DataFrame(list(zip(lista_municipio, lista_votos)),
               columns =['Municipio', 'Votos_totales'])
votos_antioquia.sort_values(by=["Votos_totales"],ascending=False)
```

```python
porcentaje = pd.merge(votos_antioquia,votos_conservador,how="right")
porcentaje
```

```python
porcentaje
porcentaje.loc["Porcentaje_votos_conservador"] = 0
```

```python
for i, row in porcentaje.iterrows():
    if row["Votos_totales"] != 0:
        porcentaje_total = (row["Votos_conservador"] / row["Votos_totales"]) * 100
        porcentaje.loc[i, "Porcentaje_votos_conservador"] = porcentaje_total
    else:
        porcentaje.loc[i, "Porcentaje_votos_conservador"] = "Indefinido"
porcentaje
porcentaje.replace({"Indefinido":0},inplace=True)
porcentaje
```

```python
porcentaje.sort_values(by="Porcentaje_votos_conservador",ascending=False)
```

```python

antioquia_verde = antioquia[antioquia["codigo_partido"]=="Partido verde"]
lista_municipio = []
lista_votos = []
for group, frame in antioquia_verde.groupby("municipio"):
    votos = np.sum(frame["votos"])
    lista_municipio.append(group)
    lista_votos.append(votos)

votos_verde = pd.DataFrame(list(zip(lista_municipio, lista_votos)),
               columns =['Municipio', 'Votos_verde'])
votos_verde.sort_values(by=["Votos_verde"],ascending=False)


porcentaje = pd.merge(porcentaje,votos_verde)

porcentaje.loc["Porcentaje_votos_verde"] = 0

for i, row in porcentaje.iterrows():
    if row["Votos_totales"] != 0:
        porcentaje_total = (row["Votos_verde"] / row["Votos_totales"]) * 100
        porcentaje.loc[i, "Porcentaje_votos_verde"] = porcentaje_total
    else:
        porcentaje.loc[i, "Porcentaje_votos_verde"] = "Indefinido"
porcentaje.replace({"Indefinido":0},inplace=True)

porcentaje.sort_values(by="Porcentaje_votos_verde",ascending=False)
```

```python
porcentaje.to_csv("porcentaje.csv")
```
